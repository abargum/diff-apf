# 1. Training

### Define Allpass Model ###

```python
ALLPASS_STRUCTURE = "Warped"

if ALLPASS_STRUCTURE == "Warped":
    from filters.Allpass_2ndOrder_W import AllPass2ndOrder_W
    from filters.Allpass_1stOrder_W import AllPass1stOrder_W
    model = torch.nn.Sequential(  
                AllPass2ndOrder_W(), 
                AllPass2ndOrder_W(), 
                AllPass2ndOrder_W(),
                AllPass1stOrder_W()).to(device)
    
elif ALLPASS_STRUCTURE == "Normal":
    from filters.Allpass_2ndOrder import AllPass2ndOrder
    from filters.Allpass_1stOrder import AllPass1stOrder
    model = torch.nn.Sequential(  
                AllPass2ndOrder(), 
                AllPass2ndOrder(), 
                AllPass2ndOrder(),
                AllPass1stOrder()).to(device)
    
else:
    from filters.Allpass_2ndOrder_SW import AllPass2ndOrder_SW
    from filters.Allpass_1stOrder_SW import AllPass1stOrder_SW
    model = torch.nn.Sequential(  
                AllPass2ndOrder_SW(), 
                AllPass2ndOrder_SW(), 
                AllPass2ndOrder_SW(),
                AllPass1stOrder_SW()).to(device)
    
n_params = 0
for name, param in model.named_parameters():
        n_params += np.prod(param.size())
print(f'The model contains a total of {n_params:,} parameters.')
```

### Generate Data Batches ###

```python
from torch.utils.data import Dataset
from torch.utils.data import DataLoader

class DIIRDataSet(Dataset):
    def __init__(self, input, target, sequence_length):
        self.input = input
        self.target = target
        self._sequence_length = sequence_length
        self.input_sequence = self.wrap_to_sequences(self.input, self._sequence_length)
        self.target_sequence = self.wrap_to_sequences(self.target, self._sequence_length)
        self._len = self.input_sequence.shape[0]

    def __len__(self):
        return self._len

    def __getitem__(self, index):
        return {'input': self.input_sequence[index, :, :]
               ,'target': self.target_sequence[index, :, :]}

    def wrap_to_sequences(self, data, sequence_length):
        num_sequences = int(np.floor(data.shape[0] / sequence_length))
        truncated_data = data[0:(num_sequences * sequence_length)]
        wrapped_data = truncated_data.reshape((num_sequences, sequence_length, 1))
        return np.float64(wrapped_data)

batch_size = 512
sequence_length = 2048

loader = DataLoader(dataset=DIIRDataSet(train_input, train_target, sequence_length), batch_size=batch_size, shuffle = True)
print("Batch dim:", next(iter(loader))['input'].size())
```

### Define Loss Function, Hyperparameters and Training Loop ###

```python
import losses
from torch.optim import Adam

folder = Path('runs1', EFFECT_TYPE)
writer = SummaryWriter(folder)

LOSS_TERM = 'M-STFT'
criterion = losses.LossModule(loss_term=LOSS_TERM).to(device)

n_epochs = 500
lr = 1e-5
optimizer = Adam(model.parameters(), lr=lr, betas=(0.9, 0.999), eps=1e-08, weight_decay=0, amsgrad=False)

losses = []

def train(criterion, model, loader, optimizer, steps):
    model.train()
    device = next(model.parameters()).device
    total_loss = 0

    for batch in loader:
        input_seq_batch = batch['input'].to(device)
        input_seq_batch = torch.nn.functional.normalize(input_seq_batch)
        target_seq_batch = batch['target'].to(device)
        
        predicted_output = model(input_seq_batch)
                
        optimizer.zero_grad()
        loss = criterion(target_seq_batch, predicted_output) 
        loss.backward()
        optimizer.step()
        
        writer.add_scalar("loss/loss pr. step", loss.item(), steps)
        steps += 1
        
        total_loss += loss.item()

    total_loss /= len(loader)
    losses.append(total_loss)
    
    return total_loss
```

### Train the Model ###

```python
from tqdm import tqdm

steps = 0

for epoch in tqdm(range(n_epochs)):
    loss = train(criterion, model, loader, optimizer, steps)
    writer.add_scalar("loss/loss pr. epoch", loss, epoch)
    if epoch % 10 == 0: 
        torch.save(model.state_dict(), f"pretrained/{EFFECT_TYPE}.pth")
```

### Inspect Learned Phase Shift
Retrieve the learned coefficients and initialise "traditional" APF pipeline for efficiency

```python
c = []
d = []
a = []

c.append([float(model[0].cell.c.data.cpu().numpy()), float(model[1].cell.c.data.cpu().numpy()), float(model[2].cell.c.data.cpu().numpy())])
d.append([float(model[0].cell.d.data.cpu().numpy()), float(model[1].cell.d.data.cpu().numpy()), float(model[2].cell.d.data.cpu().numpy())])
a.append([float(model[0].cell.a.data.cpu().numpy()), float(model[1].cell.a.data.cpu().numpy()), float(model[2].cell.a.data.cpu().numpy())])

h = float(model[3].cell.a.data.cpu().numpy())
g = float(model[3].cell.g.data.cpu().numpy())

print("C coeffs", c)
print("D coeffs", d)
print("A coeffs", a)
print( "G, H coeffs:", g, h)
```

```python
def APF1_W(input, h, g):
    state1 = 0
    output = np.zeros(len(input))
    for n in range(len(input)):  
        output[n] = ( (h + g) * input[n] + state1) / (1.0 + h * g)
        state1 = (1.0 + h * g) * input[n] - (h + g) * output[n] 
    return output

def APF2_W(input, c, d, a):
    a1 = a
    a2 = a
    state1 = 0
    state2 = 0
    output = np.zeros(len(input))
    for n in range(len(input)):
        output[n] = (input[n] * (c + a2 * a1 + a2 * d) + state1) / (1.0 + a2 * a1 * c + a2 * d)  
        state1 = input[n] * (a2 + a1 + d + a2 * c) + output[n] * (-a2 * c - a1*c - d - a2) + (a1 * (input[n] - c * output[n])) * (-a2 * a1) + (-a2 * a1 + 1.0) * state2
        state2 = input[n] - c * output[n] - a1 * (a1 * (input[n] - c * output[n]) + state2) 
    return output
```

```python
apf_output = APF2_W(train_input, c[-1][0], d[-1][0], a[-1][0])
apf_output = APF2_W(apf_output, c[-1][1], d[-1][1], a[-1][1])
apf_output = APF2_W(apf_output, c[-1][2], d[-1][2], a[-1][2])
apf_output = APF1_W(apf_output, h, g)
```
